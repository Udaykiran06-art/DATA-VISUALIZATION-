```python
#create a simple line chart and title
import matplotlib.pyplot as plt
x = [5,6,9]
y = [5,6,20]
plt.plot(x,y,marker='*')
plt.title('Boss it is a line graph')
plt.show()
```

```python
#create a simple line chart and giving title and labelling
import matplotlib.pyplot as plt
x = [5,6,9]
y = [5,6,20]
plt.title("A line graph")

plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y,marker='*')
plt.show()
```

```python
#Create two lines in single chart
import matplotlib.pyplot as plt
x = [5,6,9]
y = [5,6,20]
p = [10,20,25]
plt.title('A line graph')
plt.xlabel("X values")
plt.ylabel("Y values")

plt.plot(x,y,marker='*',color='green')
plt.plot(x,p,marker='*')
plt.show()
```

```python
#Creating Bar Chart
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales = [23,45,56,78,213,45,78,89,99,100,101,130]
colors = ['purple','green','k','blue','violet','k','green','y','pink','blue','red','green']
plt.title("A Bar Graph")

plt.xlabel("Months",color='green')

plt.ylabel("Product Sales",color='red')

plt.bar(months,sales,color=colors)

plt.show()
```

```python
#Creating horizontal bar chart
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

sales = [23,45,56,78,213,45,78,89,99,100,101,130]

colors = ['purple','green','k','blue','violet','k','green','y','pink','blue','red','green']
plt.title("A Bar Graph")

plt.xlabel("Product Sales",color='green')

plt.ylabel("Months",color='red')

plt.barh(months,sales,color=colors)

plt.show()
```

```python
#Creating Bar Chart
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales = [23,45,56,78,213,45,78,89,99,100,101,130]
colors = ['purple','green','k','blue','violet','k','green','y','pink','blue','red','green']
plt.title("A Bar Graph")

plt.xlabel("Months",color='green')

plt.ylabel("Product Sales",color='red')

plt.bar(months,sales,color=colors,width=1.5)

plt.show()
```

```python
#Creating a Histogram
import matplotlib.pyplot as plt
data = [12,15,13,20,19,20,11,19,11,12,19,13,15,16,18,13]

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Histogram Plot")
plt.hist(data,bins=20)
plt.show()
```

```python
#Creating pie chart
import matplotlib.pyplot as plt
students = ['Daniel','Prasad', 'Srinu']
points = [62, 48, 36]
colors = ['pink','green','k']

plt.pie(points,labels=students,colors=colors)
plt.show()
```

```python
#Creating pie chart
import matplotlib.pyplot as plt
students = ['Daniel','Prasad','Srinu']
points = [62,48,36]
colors = ['violet','red','blue']
plt.pie(points,labels=students,colors=colors,shadow=True,
explode=(0.05,0.05,0.05),autopct='%1.1f%%')

plt.axis('equal')
plt.legend()
plt.show()
```

```python
#Creating box plot
import matplotlib.pyplot as plt
import pandas as pd
d = {
    "Apple":[10,20,30,40],
    "Orange":[7,14,21,28],
    "Banana":[55,15,8,12],
    "Pear":[15,14,1,8]
}

i = ['Basket1','Basket2','Basket3','Basket4']

df = pd.DataFrame(d,index =i)
plt.imshow(df,cmap="YlGnBu")
plt.colorbar()
plt.xticks(range(len(df)),df.columns)
plt.yticks(range(len(df)),df.index)
plt.show()
```
